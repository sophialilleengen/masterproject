```python
level = 4

for halo_number in [24]:  # range(1, 31):
    halodir = basedir+"halo_{0}/".format(halo_number)
    snappath = halodir+"output/"

    for snapnr in range(127,128,1):
        print("level   : {0}".format(level))
        print("halo    : {0}".format(halo_number))
        print("snapnr  : {0}".format(snapnr))
        print("basedir : {0}".format(basedir))
        print("halodir : {0}".format(halodir))
        print("snappath: {0}\n".format(snappath))
        s, sf = eat_snap_and_fof(level, halo_number, snapnr, snappath, loadonlytype=[4], 
            haloid=0, galradfac=0.1, verbose=True) 

        # Clean negative and zero values of gmet to avoid RuntimeErrors
        # later on (e.g. dividing by zero)
        s.data['gmet'] = np.maximum( s.data['gmet'], 1e-40 )
        
```

```python
i_ID = np.isin(s.id,survivor_ids)
```

```python
x_all_kpc = 1000. * s.pos[(s.halo == 0) * (s.subhalo==0)][:,2]
y_all_kpc = 1000. * s.pos[(s.halo == 0) * (s.subhalo==0)][:,1]
z_all_kpc = 1000. * s.pos[(s.halo == 0) * (s.subhalo==0)][:,0]
r_all_kpc = 1000. * s.r()[(s.halo == 0) * (s.subhalo==0)]

x_plot = x_all_kpc[r_all_kpc < 1000. * s.galrad]
y_plot = y_all_kpc[r_all_kpc < 1000. * s.galrad]
z_plot = z_all_kpc[r_all_kpc < 1000. * s.galrad]
r_plot = r_all_kpc[r_all_kpc < 1000. * s.galrad]
```

```python
pos_vec = s.pos[i_ID]
```

```python
z_kpc = 1000 * pos_vec[:,0]
y_kpc = 1000 * pos_vec[:,1]
x_kpc = 1000 * pos_vec[:,2]
```

```python
cond0 = np.where((jR_kpckms >590.) & (jR_kpckms < 610.))
cond1 = np.where((jR_kpckms >1590.) & (jR_kpckms < 1610.))
cond2 = np.where((jR_kpckms >2290.) & (jR_kpckms < 2310.))

cond3 = np.where((lz_kpckms > -6470.) & (lz_kpckms < -6450.))
cond4 = np.where((lz_kpckms > -5470.) & (lz_kpckms < -5450.))
cond5 = np.where((lz_kpckms > -4470.) & (lz_kpckms < -4450.))

cond6 = np.where((jz_kpckms >360.) & (jz_kpckms < 380.))
cond7 = np.where((jz_kpckms >620.) & (jz_kpckms < 640.))
cond8 = np.where((jz_kpckms >1040.) & (jz_kpckms < 1060.))

conds = {'cond0': cond0, 'cond1': cond1, 'cond2': cond2, \
         'cond3': cond3, 'cond4': cond4, 'cond5': cond5, \
         'cond6': cond6, 'cond7': cond7, 'cond8': cond8}

acts = {'act0': 'jR', 'act1': 'jR', 'act2': 'jR',\
        'act3': 'lz', 'act4': 'lz', 'act5': 'lz',\
        'act6': 'jz', 'act7': 'jz', 'act8': 'jz'}
```

```python
cmap = copy.copy(plt.cm.Greys)
#cmap.set_bad((0,0,0))  # Fill background with black

import seaborn as sns
palette = sns.color_palette(None, 50)

for i in range(9):
    i_cond = conds['cond{}'.format(i)]
    x = x_kpc[i_cond]
    y = y_kpc[i_cond]
    z = z_kpc[i_cond]
    r = r_kpc[i_cond]
    act = acts['act{}'.format(i)]
    if act == 'jR':
        act_min = np.min(jR_kpckms[i_cond])
        act_max = np.max(jR_kpckms[i_cond])
    elif act == 'lz':
        act_min = np.min(lz_kpckms[i_cond])
        act_max = np.max(lz_kpckms[i_cond])
    elif act == 'jz':
        act_min = np.min(jz_kpckms[i_cond])
        act_max = np.max(jz_kpckms[i_cond])
    else:
        print(act + " seems to be wrong. Check!")
        
    title = "{:.1f} kpc <= {} <= {:.1f} kpc".format(act_min, act, act_max)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    h1 = ax1.hist2d(x_plot, y_plot, 201, norm = LogNorm(), cmap = cmap)
    ax1.set_title(title)
    h2 = ax2.hist2d(r_plot, z_plot, 201, norm = LogNorm(), cmap = cmap)
    for ii in range(len(x)):
        ax1.plot(x[ii], y[ii], '.', color = palette[ii])    
        ax2.plot(r[ii], z[ii], '.', color = palette[ii])
    ax1.set_xlabel("x [kpc]")
    ax1.set_ylabel("y [kpc]")
    ax1.set_aspect('equal')
    ax2.set_xlabel("r [kpc]")
    ax2.set_ylabel("z [kpc]")
    ax2.set_aspect('equal')

    plt.show()
```
