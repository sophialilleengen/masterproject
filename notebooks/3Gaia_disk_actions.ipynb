```python
#_____import packages______
from astropy.io import fits
import numpy as np
import math
import time
import sys
from astropy.table import Table
from astropy import constants as const
from astropy import units as u    
from galpy.util import bovy_coords, bovy_conversion
from galpy.potential import MWPotential2014
from galpy.actionAngle import estimateDeltaStaeckel, actionAngleStaeckel, UnboundError
import random 
import corner
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib import rcParams 
rcParams['font.size'] = 16
```

```python
#_____read in data_____
hdul = fits.open('../../Gaia/Gaia_data_for_Sophia_temp.fits')
tbdata = hdul[1].data
```

```python
random_index = tbdata.field('random_index')  
vx_HC_kms = tbdata.field('U_HC_kms')
vy_HC_kms = tbdata.field('V_HC_kms')
vz_HC_kms = tbdata.field('W_HC_kms')
U_LSR_kms = tbdata.field('U_LSR_kms')
V_LSR_kms = tbdata.field('V_LSR_kms')
W_LSR_kms = tbdata.field('W_LSR_kms')
R_kpc = tbdata.field('R_GC_kpc')
phi_rad = tbdata.field('phi_GC_rad')
z_kpc = tbdata.field('z_GC_kpc')
vR_kms = tbdata.field('vR_GC_kms')
vT_kms = tbdata.field('vT_GC_kms')
vz_kms = tbdata.field('vz_GC_kms')
```

```python
# velocity cut to distinguish disk stars (subtract v_LSR from v_T)
v_GC_cut = np.sqrt(vR_kms**2 + (vT_kms-232.)**2 + vz_kms**2)
disk_v_cut = (v_GC_cut < 210.)

halo_v_cut = (v_GC_cut > 210.)
```

```python
toomre_x_galcen = vT_kms
toomre_y_galcen = np.sqrt(vR_kms**2 + vz_kms**2)


cut = (toomre_x_galcen < 300.) * (toomre_y_galcen < 500.)
fig,ax = plt.subplots()
ax.spines['top'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.xaxis.set_tick_params(width=1.5)
ax.yaxis.set_tick_params(width=1.5)

ax.hist2d(toomre_x_galcen[cut], toomre_y_galcen[cut], bins = 201, norm = colors.LogNorm(), cmap = plt.cm.binary)
ax.plot(toomre_x_galcen[disk_v_cut], toomre_y_galcen[disk_v_cut], 'r.', alpha = 0.1)
ax.set_xlabel('v$_\mathrm{T}$ (km s$^{-1}$)')
ax.set_ylabel('$\sqrt{v_R^2 + v_z^2}$ (km s$^{-1}$)')

plt.show()
```

```python
fix_R0_kpc = 8.
fix_v0_kms = 220.

R_galpy, vR_galpy, vT_galpy, z_galpy, vz_galpy = R_kpc[disk_v_cut] / fix_R0_kpc, vR_kms[disk_v_cut] / fix_v0_kms, vT_kms[disk_v_cut] / fix_v0_kms, z_kpc[disk_v_cut] / fix_R0_kpc, vz_kms[disk_v_cut] / fix_v0_kms
randint = random.sample(range(len(R_galpy)), 50000)
delta = estimateDeltaStaeckel(MWPotential2014,R_galpy[randint],z_galpy[randint])
print('delta = {:3f}; therefore we take delta = 0.62'.format(delta))

aAS= actionAngleStaeckel(pot=MWPotential2014,delta=0.62)
jR_galpy, lz_galpy, jz_galpy = aAS(R_galpy[randint], vR_galpy[randint], vT_galpy[randint], z_galpy[randint], vz_galpy[randint])


jR_kpckms = jR_galpy[np.isfinite(jR_galpy)] * fix_R0_kpc * fix_v0_kms 
lz_kpckms = lz_galpy[np.isfinite(lz_galpy)] * fix_R0_kpc * fix_v0_kms 
jz_kpckms = jz_galpy[np.isfinite(jz_galpy)] * fix_R0_kpc * fix_v0_kms 
np.savetxt('../files/Gaia_disk_actions_MW14.txt', (jR_kpckms, lz_kpckms, jz_kpckms), header = 'jR [kpc km/s]\tlz [kpc km/s]\tjz [kpc km/s]')
data_cut = (jR_kpckms<=3000. )* (jz_kpckms<= 500)
data = np.vstack([jR_kpckms[data_cut], lz_kpckms[data_cut], jz_kpckms[data_cut]])
labels = ['jR [kpc km s$-1$]', 'lz [kpc km s$-1$]', 'jz [kpc km s$-1$]']
figure = corner.corner(data.transpose(), labels = labels, plot_contours = 1, color = 'black')            
figure.savefig('../plots/Gaia_disk_actions_MW14.png', dpi = 300, bbox_inches='tight')
```

```python
fix_R0_kpc = 8.
fix_v0_kms = 220.

R_galpy, vR_galpy, vT_galpy, z_galpy, vz_galpy = R_kpc[halo_v_cut] / fix_R0_kpc, vR_kms[halo_v_cut] / fix_v0_kms, vT_kms[halo_v_cut] / fix_v0_kms, z_kpc[halo_v_cut] / fix_R0_kpc, vz_kms[halo_v_cut] / fix_v0_kms
delta = estimateDeltaStaeckel(MWPotential2014,R_galpy,z_galpy)
print('delta = {:3f}; therefore we take delta = 0.45'.format(delta))
aAS= actionAngleStaeckel(pot=MWPotential2014,delta=0.45)

# some orbit particles seem to be unbound
jR_galpy = np.full(len(R_galpy), np.nan)
lz_galpy = np.full(len(R_galpy), np.nan)
jz_galpy = np.full(len(R_galpy), np.nan)

for ii in range(len(R_galpy)):
    try: 
        jR_galpy[ii], lz_galpy[ii], jz_galpy[ii] = aAS(R_galpy[ii], vR_galpy[ii], vT_galpy[ii], z_galpy[ii], vz_galpy[ii])

    except(ValueError, UnboundError): continue

jR_kpckms = jR_galpy[np.isfinite(jR_galpy)] * fix_R0_kpc * fix_v0_kms 
lz_kpckms = lz_galpy[np.isfinite(lz_galpy)] * fix_R0_kpc * fix_v0_kms 
jz_kpckms = jz_galpy[np.isfinite(jz_galpy)] * fix_R0_kpc * fix_v0_kms 

data_cut = (jR_kpckms<=3000. )* (jz_kpckms<= 500)
data = np.vstack([jR_kpckms[data_cut], lz_kpckms[data_cut], jz_kpckms[data_cut]])
labels = ['jR [kpc km s$-1$]', 'lz [kpc km s$-1$]', 'jz [kpc km s$-1$]']
figure = corner.corner(data.transpose(), labels = labels, plot_contours = 1, color = 'black')            
figure.savefig('../plots/Gaia_halo_actions_MW14.png', dpi = 300, bbox_inches='tight')
```
