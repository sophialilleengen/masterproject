```python
#### path = /hits/universe/GigaGalaxy/level4_MHD/halo_24/output/*
level = 4
prognum = 1
prog_IDs = np.loadtxt(filedir + 'merger_trees/progenitor{}_IDs'.format(prognum))

for halo_number in [24]:  # range(1, 31):
    halodir = basedir+"halo_{0}/".format(halo_number)
    snappath = halodir+"output/"
    for snapnr in range(60,128,1):
        print("level   : {0}".format(level))
        print("halo    : {0}".format(halo_number))
        print("snapnr  : {0}".format(snapnr))
        print("basedir : {0}".format(basedir))
        print("halodir : {0}".format(halodir))
        print("snappath: {0}\n".format(snappath))
        s, sf = eat_snap_and_fof(level, halo_number, snapnr, snappath, loadonlytype=[4], 
            haloid=0, galradfac=0.1, verbose=True) 

        # Clean negative and zero values of gmet to avoid RuntimeErrors
        # later on (e.g. dividing by zero)
        s.data['gmet'] = np.maximum( s.data['gmet'], 1e-40 )

        gcmask = np.isin(s.id, prog_IDs)

        if np.sum(gcmask) == 0:
            print('In snapshot {}, no particles or progenitor {} are found.'.format(snapnr, prognum))

        ### plot stars of halo 0 in Simulation 24
        istars, = np.where( (s.type == 4) & (s.halo == 0))#& (s.subhalo == 0))
        (x, y, z), (vx, vy, vz), rxyz, rxy = get_cartesian_vectors(s, sf, istars)

        binnum = 401
        H1, xedges1, yedges1 = np.histogram2d(x, y, bins=binnum)
        H2, xedges2, yedges2 = np.histogram2d(x, z, bins=binnum)
        vmin = 1e0
        if np.max(H1) > np.max(H2):
            vmax = np.max(H1)
        else: vmax = np.max(H2)

        fig, ax = plt.subplots()
        h = ax.hist2d(x, y, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax = vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("y [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_halo0/xy_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()

        fig, ax = plt.subplots()
        h = ax.hist2d(x, z, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax= vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("z [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_halo0/xz_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
        
        ### plot stars of subhalo 0 
        istars, = np.where( (s.type == 4) & (s.halo == 0) & (s.subhalo == 0 ))#& (s.subhalo == 0))
        (x, y, z), (vx, vy, vz), rxyz, rxy = get_cartesian_vectors(s, sf, istars)

        binnum = 401
        H1, xedges1, yedges1 = np.histogram2d(x, y, bins=binnum)
        H2, xedges2, yedges2 = np.histogram2d(x, z, bins=binnum)
        vmin = 1e0
        if np.max(H1) > np.max(H2):
            vmax = np.max(H1)
        else: vmax = np.max(H2)

        fig, ax = plt.subplots()
        h = ax.hist2d(x, y, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax = vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("y [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_subhalo0/xy_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
        

        fig, ax = plt.subplots()
        h = ax.hist2d(x, z, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax= vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("z [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_subhalo0/xz_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()

        ### plot stars of halo 0 in Simulation 24 with marked progenitors
        istars, = np.where( (s.type == 4) & (s.halo == 0))
        (x, y, z), (vx, vy, vz), rxyz, rxy = get_cartesian_vectors(s, sf, istars)
        (x_gcs, y_gcs, z_gcs), (vx_gcs, vy_gcs, vz_gcs), rxyz_gcs, rxy_gcs = get_cartesian_vectors(s, sf, gcmask)
        
        binnum = 101
        H1, xedges1, yedges1 = np.histogram2d(x, y, bins=binnum)
        H2, xedges2, yedges2 = np.histogram2d(x, z, bins=binnum)
        
        H1_gcs, xedges1_gcs, yedges1_gcs = np.histogram2d(x_gcs, y_gcs, bins=binnum)
        H2_gcs, xedges2_gcs, yedges2_gcs = np.histogram2d(x_gcs, z_gcs, bins=binnum)
        vmin = 1e0
        if np.max(H1) > np.max(H2):
            vmax = np.max(H1)
        else: vmax = np.max(H2)

        fig, ax = plt.subplots()
        h = ax.hist2d(x, y, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax = vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("y [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_halo0/xy_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
        
        fig, ax = plt.subplots()
        h = ax.hist2d(x, z, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax= vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("z [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_halo0/xz_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
        
        
        ### plot stars of subhalo 0 with marked progenitors
        istars, = np.where( (s.type == 4) & (s.halo == 0) & (s.subhalo == 0 ))
        (x, y, z), (vx, vy, vz), rxyz, rxy = get_cartesian_vectors(s, sf, istars)
        (x_gcs, y_gcs, z_gcs), (vx_gcs, vy_gcs, vz_gcs), rxyz_gcs, rxy_gcs = get_cartesian_vectors(s, sf, gcmask)

        binnum = 101
        H1, xedges1, yedges1 = np.histogram2d(x, y, bins=binnum)
        H2, xedges2, yedges2 = np.histogram2d(x, z, bins=binnum)
        H1_gcs, xedges1_gcs, yedges1_gcs = np.histogram2d(x_gcs, y_gcs, bins=binnum)
        H2_gcs, xedges2_gcs, yedges2_gcs = np.histogram2d(x_gcs, z_gcs, bins=binnum)

        vmin = 1e0
        if np.max(H1) > np.max(H2):
            vmax = np.max(H1)
        else: vmax = np.max(H2)

        fig, ax = plt.subplots()
        h = ax.hist2d(x, y, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax = vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("y [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_subhalo0/xy_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
        

        fig, ax = plt.subplots()
        h = ax.hist2d(x, z, bins=binnum, norm=matplotlib.colors.LogNorm(), cmap =cmap, vmin = vmin, vmax= vmax)
        ax.set_xlabel("x [kpc]")
        ax.set_ylabel("z [kpc]")
        ax.set_aspect('equal')
        cbar = fig.colorbar(h[3])
        fig.savefig(plotdir + 'Auriga_overview/spatial_distribution_subhalo0/xz_dist_snap_{}.png'.format(snapnr), format = 'png', dpi = 300, bbox_inches = 'tight')
        plt.show()
```
